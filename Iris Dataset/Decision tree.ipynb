```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import datasets 
from sklearn import tree
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report
from sklearn import preprocessing 
from sklearn.tree import DecisionTreeClassifier
```

```python
names = [' sepal length', 'sepal width' , 'petal length', 'petal width'  ,'class']
iris = pd.read_csv('/Users/users/Downloads/iris.data', names=names)
```

```python
label_encoder = preprocessing.LabelEncoder()
iris['class']= label_encoder.fit_transform(iris['class'])
```

```python
iris
```

```python
x=iris.iloc[:, 0:4]
y=iris.iloc[:, 4]
```

```python
y
```

```python
iris['class'].unique()
```

```python
iris['class'].value_counts()
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=4)
```

```python
model = DecisionTreeClassifier(criterion ='entropy', max_depth=5)
model.fit(x_train,y_train)
tree.plot_tree(model)
```

```python
fn = [' sepal length', 'sepal width' , 'petal length', 'petal width'  ]
cn= ['setoss', 'versicolour', 'virginca']
fig, axes = plt.subplots(nrows=1 , ncols=1 , figsize= (4,4), dpi=300)
tree.plot_tree(model, feature_names = fn,
              class_names=cn,
              filled = True)
```

```python
preds = model.predict(x_test)
pd.Series(preds).value_counts()
```
